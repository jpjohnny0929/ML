```
def baseline(n):
    
    b=[]
    for i in range(0,n+1):
        line=[]
        for j in range(1):
            a=df_centered.loc[i]-whittaker(i)[0]
            line.append(a)
        b.append(line)
    return b
```

```
ma=baseline(308)
man=pd.DataFrame(data=ma[0])

for i in range (1,309):
    man=man.append(ma[i])
man
```

```
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```
x=pca_df.iloc[:,0:2]

kmeans = KMeans(2)
kmeans.fit(x)
```

```
wcss=[]
for i in range(1,11):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,11)
plt.plot(number_clusters,wcss)
plt.title('Elbow Curve (Eigendecomposition)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```
identified_clusters = kmeans.fit_predict(x)

data_with_clusters = pca_df
plt.scatter(data_with_clusters['PC1'],data_with_clusters['PC2'],c=identified_clusters,cmap='rainbow')
plt.title('K-means Clustering (Eigendecomposition)')
```

```
def whittaker_asls(n):
    return pybaselines.whittaker.asls(df_centered.loc[n], lam=1000000.0, p=0.01, diff_order=1, max_iter=50, tol=0.001, weights=None)
c=df_centered.loc[10]-whittaker_asls(10)[0]
df_centered.loc[10].plot()
c.plot()
c
```

```
def modpoly(n):
    return pybaselines.polynomial.modpoly(df_centered.loc[n], x_data=None, poly_order=2, tol=0.001, max_iter=250, weights=None, use_original=False, mask_initial_peaks=False, return_coef=False)
c=df_centered.loc[10]-modpoly(10)[0]
df_centered.loc[10].plot()
c.plot()
c
```
