```
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

path='https://raw.githubusercontent.com/jpjohnny0929/ML/main/anonymized_data.csv'
df= pd.read_csv(path)
df
```

```
a=df.to_numpy().max()
b=df.to_numpy().min()
print (a, b)
```

```
def spectrum_plot(n):
    
#plot the spectrum of the first "n" datapoints 

    for i in range(0,n+1):
        df.loc[i].plot()
        
    return "spectrum for first {0} datapoints".format(n)

spectrum_plot(308)
plt.title('Raw Spectra')
```

```
df_center = df.T.apply(lambda x: x-x.mean())
df_centered=df_center.T
df_centered
```

```
df_centered.T.plot(legend=None)
plt.title('Mean-centered Spectra')
```

```
import pybaselines
from pybaselines import utils

def whittaker(n):
    return pybaselines.whittaker.arpls(df_centered.loc[n], lam=100000.0, diff_order=1, max_iter=50, tol=0.001, weights=None)

a=whittaker(10)
print(len(a[0]))
```

```
def baseline(n):
    
    b=[]
    for i in range(0,n+1):
        line=[]
        for j in range(1):
            a=df_centered.loc[i]-whittaker(i)[0]
            line.append(a)
        b.append(line)
    return b
```

```
ma=baseline(308)
center_base=pd.DataFrame(data=ma[0])

for i in range (1,309):
    center_base=center_base.append(ma[i])
    
center_base
```

```
center_base.T.plot(legend=None)
plt.title('Mean-centered and Baseline-corrected Spectra')
```

```
pca=PCA()
pca.fit(center_base)
pca_data=pca.transform(center_base)
```

```
per_var=np.round(pca.explained_variance_ratio_ *100, decimals=1)
labels=['PC'+str(x) for x in range(1,11)]

plt.bar(x=range(1,11), height=per_var[0:10], tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot (Eigendecomposition)')
plt.show()
```

```
per_var=np.round(pca.explained_variance_ratio_ *100, decimals=1)
print(per_var[0:10])
print(len(per_var))
```

```
labels=['PC'+str(x) for x in range(1,len(per_var)+1)]
pca_df=pd.DataFrame(pca_data, columns=labels)
pca_df
```

```
plt.scatter(pca_df.PC1, pca_df.PC2)
plt.title('PCA with Eigendecomposition')
plt.xlabel('PC1-{0}%'.format(per_var[0]))
plt.ylabel('PC2-{0}%'.format(per_var[1]))

    
plt.show()
```

```
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```
x=pca_df.iloc[:,0:2]

kmeans = KMeans(6)
kmeans.fit(x)
```

```
wcss=[]
for i in range(1,11):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,11)
plt.plot(number_clusters,wcss)
plt.title('Elbow Curve (Eigendecomposition)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```
identified_clusters = kmeans.fit_predict(x)

data_with_clusters = pca_df
plt.scatter(data_with_clusters['PC1'],data_with_clusters['PC2'],c=identified_clusters,cmap='rainbow')
plt.title('K-means Clustering (Eigendecomposition)')
```
