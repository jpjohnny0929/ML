```
import pandas as pd
import numpy as np

path='https://raw.githubusercontent.com/jpjohnny0929/ML/main/anonymized_data.csv'
df= pd.read_csv(path)
df

df_center = df.T.apply(lambda x: x-x.mean())
df_centered=df_center.T
df_centered
```

```
import pybaselines
from pybaselines import utils

def whittaker(n):
    return pybaselines.whittaker.arpls(df_centered.loc[n], lam=100000.0, diff_order=1, max_iter=50, tol=0.001, weights=None)

a=whittaker(10)
print(len(a[0]))
```

```
def baseline(n):
    
    b=[]
    for i in range(0,n+1):
        line=[]
        for j in range(1):
            a=df_centered.loc[i]-whittaker(i)[0]
            line.append(a)
        b.append(line)
    return b
```

```
ma=baseline(308)
center_base=pd.DataFrame(data=ma[0])

for i in range (1,309):
    center_base=center_base.append(ma[i])
    
center_base
```
